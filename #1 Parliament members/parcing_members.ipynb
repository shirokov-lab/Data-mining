```python
# import libs
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json
```

```python
# parcing every url(member-card) into list
persons_url_list = []
for i in range(0,740,12):
    url = f'https://www.bundestag.de/ajax/filterlist/en/members/863330-863330?limit=12&noFilterSet=true&offset={i}'
    q = requests.get(url)
    result = q.content
    soup = BeautifulSoup(result,'lxml')
    persons = soup.find_all('a')
    for person in persons:
        person_page_url =  person.get('href')
        persons_url_list.append(person_page_url)
```

```python
# save list of urls as txt 
with open('persons_url_list.txt','w') as file:
    for line in persons_url_list:
        file.write(f'{line}\n')
```

```python
# check txt file
display(pd.read_csv("persons_url_list.txt").head(2))
display(pd.read_csv("persons_url_list.txt").shape)
```

```python
# load txt file of urls
with open('persons_url_list.txt') as file:
    lines = [line.strip() for line in file.readlines()]
```

```python
# get Name, Company name and Social networks from 1 url
line = lines[0]

q = requests.get(line)
result = q.content
soup = BeautifulSoup(result,'lxml')

person_name_company = soup.find(class_='col-xs-8 col-md-9 bt-biografie-name').find('h3').text.strip().split(',')
person_name = person_name_company[0].strip()
person_company = person_name_company[1].strip()

print(person_name)
print(person_company)

social_networks = soup.find_all(class_='bt-link-extern')
social_networks_urls = []
for item in social_networks:
    social_networks_urls.append(item.get('href'))
print(social_networks_urls)
```

```python
# get from all and save to json
data_list = []
counter = 0
for line in lines:
    q = requests.get(line)
    result = q.content
    soup = BeautifulSoup(result,'lxml')
    
    person_name_company = soup.find(class_='col-xs-8 col-md-9 bt-biografie-name').find('h3').text.strip().split(',')
    person_name = person_name_company[0].strip()
    person_company = person_name_company[1].strip()
    
    social_networks = soup.find_all(class_='bt-link-extern')
    social_networks_urls = []
    for item in social_networks:
        social_networks_urls.append(item.get('href'))
    
    
    
    data_dict = {
        'member_name':person_name,
        'company_name':person_company,
        'networks_links':social_networks_urls      
    }
    data_list.append(data_dict)
    with open('data.json','w') as json_file:
        json.dump(data_list, json_file, indent=4)

    counter+=1
    print(f'{counter/740*100}%')
```
